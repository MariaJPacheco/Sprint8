Dos variables numéricas


```python
cantidad_productos = tabla_conexion.groupby('id').size().reset_index(name='cantidad_productos')
tablas_unidas = transactions.merge(cantidad_productos, on='id', how='left')

sns.lineplot(data=tablas_unidas, x='cantidad_productos', y='amount', color='#FF00FF')
plt.title("Monto de transacciones por cantidad de productos", fontsize=18)
plt.xlabel("Cantidad de productos por transacción", fontsize=12)
plt.ylabel("Monto medio de la transacción", fontsize=12)
plt.gca().xaxis.set_major_locator(mticker.MultipleLocator(1))
plt.show()

#Aquí usé las variables numéricas: 1) 'amount' 2)'cantidad_productos', esta última calculada por mí
```

📊 Grafiqué estos datos para  visualizar el monto de las transacciones según la cantidad de productos por compra, que va de 1 a 4 productos. El lineplot muestra un rango de precios similar en todas las categorías, pero hay diferencias interesantes en la media por lo siguiente:
- Las transacciones de 1 y 3 productos tienen un monto medio más alto, alrededor de 270 euros.
- Las compras con 2 productos presentan el monto medio menor, alrededor de 245 euros.
- Las comrpas con 4 productos presentan una media intermedia, alrededo de 250 euros.

Esto siguiente que las transacciones del primer rango (1 y 3) tienden a incluír artículos de mayor valor relativo, mientras que las compras con 2 productos podrían centrarse en artículos de menor valor, lo cual refleja el comportamiento de compra de nuestros usuarios, sobre lo cual se podrían aplicar diferentes campañas para elevar el monto de las transacciones con 2 productos (ofreciendo por ejemplo packs o promociones)

➖ Ejercicio 1
Correlación de todas las variables numéricas

```python
#Primero he unido las tablas para poder tener los datos juntos.
correlacion = pd.merge(transactions, tabla_conexion, on='id', how='left')
correlacion = pd.merge(correlacion, products, left_on='product_ids_y', right_on='id', how='left')
correlacion['declined'] = correlacion['declined'].astype('str')
```

```python
#Luego hago la correlación y el mapa de calor
correlacion2 = correlacion.corr(numeric_only=True, method='spearman')
sns.heatmap(correlacion2, cmap="coolwarm", center=0)
plt.title("Correlación entre las variables numéricas", fontsize=18)
plt.show()
```

📊Este mapa de calor muestra las correlaciones entre las variables amount, price y weight. Se observa una correlación positiva alta entre price y weight, lo que indica que a medida que aumenta el peso, el precio también tiende a incrementarse de forma proporcional. Por otro lado, amount no presenta una relación lineal significativa ni con el precio ni con el peso, ya que sus correlaciones son cercanas a 0. En general, el análisis sugiere que el peso influye directamente en el precio, mientras que el monto de las transacciones parece independiente de estas variables.
